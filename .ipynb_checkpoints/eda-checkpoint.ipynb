```python
categorical = data[['PetType','Breed','Color','Size','Vaccinated','HealthCondition','PreviousOwner','AdoptionLikelihood','PetID']]
numerical = data.drop(columns = ['PetType','Breed','Color','Size','Vaccinated','HealthCondition','PreviousOwner','AdoptionLikelihood','PetID'])
```

# (3) Numerical Data Analysis 

# (3.1) Create Boxplots for each variable to identify if any outliers present

```python
# find if there are outliers in any numerical data

plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical.columns, 1):
    plt.subplot(len(numerical.columns), 1, i)
    sns.boxplot(x=numerical[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

Numerical Variables do not have any obvious outliers, hence no dropping of outliers is required

```python
numerical_scaled = pd.DataFrame(StandardScaler().fit_transform(numerical), columns=numerical.columns)
corr_matrix = numerical_scaled.corr()
target_corr = corr_matrix[['TimeInShelterDays']].drop(index='TimeInShelterDays')
plt.figure(figsize=(5, 5))
sns.heatmap(target_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Feature Correlation with TimeInShelterDays")
plt.show()
```
